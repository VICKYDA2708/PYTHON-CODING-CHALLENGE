```python
# Feedback Analyzer

# 1. Ask the user to input a sentence
feedback = input("Enter your feedback: ")

# 2. Count how many words are in the sentence
words = feedback.split()
word_count = len(words)

# 3. Count how many times the word "data" appears (case-insensitive)
data_count = feedback.lower().count("data")

# 4. Extract first 5 and last 5 characters
first_5 = feedback[:5]
last_5 = feedback[-5:]

# 5. Reverse the sentence
reversed_feedback = feedback[::-1]

# Display the results
print("\n--- Feedback Analysis Results ---")
print(f"Total number of words: {word_count}")
print(f"Number of times 'data' appears: {data_count}")
print(f"First 5 characters: {first_5}")
print(f"Last 5 characters: {last_5}")
print(f"Reversed feedback: {reversed_feedback}")
```
