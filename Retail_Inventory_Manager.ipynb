```python
# 1. Fixed product categories (Tuple - Immutable)
product_categories = ("Electronics", "Stationery", "Clothing")

# 2. List of products (Mutable)
products = ["Laptop", "Pen", "Shirt"]

# Display initial data
print("Initial Product Categories:", product_categories)
print("Initial Products:", products)
```

```python
# 3. Ask the user to enter a new product
new_product = input("\nEnter a new product to add: ")
products.append(new_product)
```

```python
# 4. Ask user for index of a product to remove
print("\nCurrent Products:", products)
remove_index = int(input("Enter the index of the product you want to remove: "))
```

```python
# Check if the index is valid
if 0 <= remove_index < len(products):
    removed_item = products.pop(remove_index)
    print(f"\nRemoved product: {removed_item}")
else:
    print("\nInvalid index! No product removed.")
```

```python
# 5. Final output
print("\n--- Final Inventory ---")
print("Product Categories (Tuple):", product_categories)
print("Updated Products List:", products)
```
