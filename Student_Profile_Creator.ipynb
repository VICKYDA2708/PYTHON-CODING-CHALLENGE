```python
# Student Profile Creator

# 1. Create variables for student information
name = input("Enter the student's name: ")
age = int(input("Enter the student's age: "))

# Favourite subjects stored in a list
favourite_subjects = ["Math", "Science", "English"]


```

```python
# 2. Display student information
print("\n--- Student Profile ---")
print(f"Name: {name}")
print(f"Age: {age}")
print("Favourite Subjects:", favourite_subjects)
```

```python
# 3. Ask the student to add a new favourite subject
new_subject = input("\nEnter a new favourite subject to add: ")
```

```python
# Add it to the list
favourite_subjects.append(new_subject)
```

```python
# 4. Display updated list
print("\n--- Updated Student Profile ---")
print(f"Name: {name}")
print(f"Age: {age}")
print("Updated Favourite Subjects:", favourite_subjects)
```
