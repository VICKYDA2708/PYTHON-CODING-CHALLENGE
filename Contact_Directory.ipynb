```python
# Contact Directory (2D List)
# Create a 2D list for contacts (name, phone, email)
contacts = [
    ["Rahul", "9876543210", "rahul@example.com"],
    ["Ananya", "9123456780", "ananya@example.com"],
    ["Vignesh", "9000001111", "vignesh@example.com"]
]
contacts
```

```python
# Allow the user to add a new contact
name = input("Enter contact name: ")
phone = input("Enter phone number: ")
email = input("Enter email address: ")
# Append new contact
contacts.append([name, phone, email])
print("\nContact added successfully!\n")
```

```python
contacts
```

```python
#Display all contacts sorted alphabetically by name
contacts.sort(key=lambda x: x[0])   # Sort by name
print("All Contacts (Sorted Alphabetically):")
for c in contacts:
    print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")
```

```python
#Search for a contact (supports partial match)
search_term = input("\nEnter name to search: ").lower()
print("\nSearch Results:")
found = False
for c in contacts:
    if search_term in c[0].lower():  # partial match check
        found = True
        print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")

if not found:
    print("No matching contacts found.")
```

```python
# Display only the first 3 contacts after sorting
print("\nTop 3 Contacts:")
for c in contacts[:3]:    # slicing
    print(f"Name: {c[0]}, Phone: {c[1]}, Email: {c[2]}")
```
