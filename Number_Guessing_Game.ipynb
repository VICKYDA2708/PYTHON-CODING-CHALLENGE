```python
import random

#Random number between 1 and 20
secret_number = random.randint(1, 20)

attempts = 0
max_attempts = 5

print("Welcome to the Number Guessing Game!")
print("Guess a number between 1 and 20.\n")

#While loop for 5 attempts
while attempts < max_attempts:
    guess = input("Enter your guess: ")

    #Handle non-integer inputs
    if not guess.isdigit():
        print("Invalid input! Please enter a number.\n")
        continue   # Do not count this as an attempt

    guess = int(guess)
    attempts += 1

    #Compare guesses
    if guess < secret_number:
        print("Too low!\n")
    elif guess > secret_number:
        print("Too high!\n")
    else:
        print(f"üéâ Correct! You guessed the number in {attempts} tries!")
        break       # Exit the loop on correct guess

#Else runs ONLY if loop didn't break
else:
    print("\n‚ùå You used all attempts!")
    print(f"The correct number was: {secret_number}")
```
