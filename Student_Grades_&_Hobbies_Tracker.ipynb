```python
# Create a dictionary with student names and grades
grades = {
    "Rahul": 85,
    "Ananya": 92,
    "Vignesh": 78
}

print("Initial Grades:", grades)
```

```python
# Allow user to add or update grades for at least 3 students
print("\nEnter/Update grades for 3 students:")
for i in range(3):
    name = input(f"Enter student name #{i+1}: ")
    grade = int(input(f"Enter grade for {name}: "))
    grades[name] = grade  # add or update
    print(f"Grade updated for {name}!")

print("\nUpdated Grades:", grades)
```

```python
# Create a set of hobbies
hobbies = set()

print("\nEnter 3 hobbies:")
for i in range(3):
    hobby = input(f"Enter hobby #{i+1}: ")
    hobbies.add(hobby)

print("\nHobbies Set:", hobbies)
```

```python
# Convert set to frozenset
frozen_hobbies = frozenset(hobbies)

print("\nFrozen Hobbies (Immutable):")
print(frozen_hobbies)
```
